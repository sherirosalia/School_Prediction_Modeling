##### Clean Financial data

```python
#import csv
financial = pd.read_csv('money_clean.csv', index_col=0)
financial.head()

```

```python
financial.dtypes
```

```python
df = financial
```

```python
df['Grants'] = df['Grants'].apply(lambda x: float(x.split()[0].replace(',', '')))
df.head()
```

```python
df.dtypes
```

```python
df['Loans'] = df['Loans'].apply(lambda x: float(x.split()[0].replace(',', '')))
df.dtypes
```

```python
df['Federal'] = df['Federal'].apply(lambda x: float(x.split()[0].replace(',', '')))
df.dtypes
```

```python
#convert to float
df['Net_Total'] = df['Net_Total'].apply(lambda x: float(x.split()[0].replace(',', '')))
df.dtypes
```

```python
df['Institutional'] = df['Institutional'].apply(lambda x: float(x.split()[0].replace(',', '')))
df.head()





```

```python
df.dtypes
```
