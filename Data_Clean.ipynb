### Bring in first dataset - student academic term data

```python
#import general student grade and course info
student_data = pd.read_csv('grade_data.csv')
```

```python
#select datapoints
new = student_data[['Student ID', 'Letter', 'Department', "Grade", 'Term GPA','Course', 'Name', 'Age', '% Attended', \
                 'Gender','Hispanic or Latino', 'Race(s)', \
                 'Primary Faculty', 'ZIP', 'LAT', 'LNG']].copy()
```

```python
new.sample(10)
```

```python
#rename columns
new.rename(columns={'% Attended': 'Attendance', 'Department':'Department','Grade':'Grade', 'Term GPA': 'Term_GPA', "Letter": 'LtrGrade', 'Student ID': 'ID', 'Hispanic or Latino' : 'Latino', 'Race(s)' : 'Race', 'Primary Faculty': 'Professor'}, inplace=True)
new.head(25)

```

```python
# Change "Female" to 2 and Male to 1 in response columns
gender_df = new.replace({'MALE': 2, 'FEMALE': 3})
gender_df.head()
```

```python
gender_df['Gender'].value_counts()
```

```python
#drop unknown gender
gender_df.drop(gender_df.loc[new['Gender']=='UNKNOWN'].index, inplace=True)
```

```python
gender_df['Gender'].value_counts()
```

```python
new = gender_df
new.head()
```

```python
#remove gaps in Professor ino
new.drop(new.loc[new['Professor']=='0'].index, inplace=True)
new.head()
```

```python
professor_df = new
```

```python
#keep professors who have taught 30 plus classes
#professor_df.groupby(['Professor']).ID.count()[lambda x:x>=29]
```

```python
professor_df = professor_df.assign(ProfID=(professor_df['Professor']).astype('category').cat.codes)
professor_df.head()
```

```python
#assign numbers to names (privacy)

c =professor_df.Professor.astype('category')
df = dict(enumerate(c.cat.categories))
df
```
